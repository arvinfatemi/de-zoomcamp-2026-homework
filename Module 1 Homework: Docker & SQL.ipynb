#### Preparation

```python
import pandas as pd
```

```python
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2025-11.parquet')
df.head()
```

```python
zones_df = pd.read_csv('https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv')
zones_df.head()
```

```python
df_with_zones = df.merge(zones_df, left_on='PULocationID', right_on='LocationID', how='left')
df_with_zones.head()
```

```python
df.shape, df_with_zones.shape
```

#### Question 3:
 For the trips in November 2025, how many trips had a trip_distance of less than or equal to 1 mile?

```python
mask = (df['lpep_pickup_datetime'] >= '2025-11-01') & (df['lpep_pickup_datetime'] < '2025-12-01') & (df['trip_distance'] <= 1)
num_trips = df.loc[mask].shape[0]
num_trips
```

#### Question 4:
 Which was the pick up day with the longest trip distance? Only consider trips with trip_distance less than 100 miles.

```python
mask = df['trip_distance'] < 100
longest_trip = df.loc[mask].sort_values(by='trip_distance', ascending=False).iloc[0]
longest_trip_date = longest_trip['lpep_pickup_datetime'].date()
longest_trip_distance = longest_trip['trip_distance']
(longest_trip_date, longest_trip_distance)
```

#### Question 5:
 Which was the pickup zone with the largest total_amount (sum of all trips) on November 18th, 2025? 

```python
mask = (df_with_zones['lpep_pickup_datetime'] >= '2025-11-18') & (df_with_zones['lpep_pickup_datetime'] < '2025-11-19')
df_nov18 = df_with_zones.loc[mask]
total_amount_by_zone = df_nov18.groupby('Zone')['total_amount'].sum()
top_zone = total_amount_by_zone.idxmax()
top_zone
```

#### Question 6:
 For the passengers picked up in the zone named "East Harlem North" in November 2025, which was the drop off zone that had the largest tip?

```python
top_pickup_zone = 'East Harlem North'
top_pickup_zone_id = zones_df[zones_df['Zone'] == top_pickup_zone]['LocationID'].values[0]
mask = (df_with_zones['PULocationID'] == top_pickup_zone_id) & (df_with_zones['lpep_pickup_datetime'] >= '2025-11-01') & (df_with_zones['lpep_pickup_datetime'] < '2025-12-01')
df_east_harlem = df_with_zones.loc[mask]
max_tip_row = df_east_harlem.loc[df_east_harlem['tip_amount'].idxmax()]
top_dropoff_zone_id = max_tip_row['DOLocationID']
zones_df[zones_df['LocationID'] == top_dropoff_zone_id]['Zone'].values[0]
```
